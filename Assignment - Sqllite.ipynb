```python
import sqlite3
import pandas as pd
import os
from pandas import DataFrame, Series

from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

adult = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data") 
```

```python
# Naming the columns as defined
adult.columns = ['age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship','race',
    'sex','capital_gain','capital_loss','hours_per_week','native_country','class']
adult
```

```python
# Displaying the first 10 rows using the limit keyword
pysqldf("select * from adult limit 10;")
```

```python
# avg() method is used to get the average and trim function is used to remove the unwanted spaces
pysqldf("select avg(hours_per_week) from adult where trim(workclass)='Private' and trim(sex)='Male';")
```

```python
# Frequency distribution on education is obtained using the count() method applied on the grouped education column
pysqldf("select education, count(*) as Frequency from adult group by education order by count(*) desc;")
```

```python
# Frequency distribution on occupation is obtained using the count() method applied on the grouped occupation column
pysqldf("select occupation, count(*) as Frequency from adult group by occupation order by count(*) desc;")
```

```python
# Frequency distribution on relationship is obtained using the count() method applied on the grouped relationship column
pysqldf("select relationship, count(*) as Frequency from adult group by relationship order by count(*) desc;")
```

```python
# Below sql provides the number of persons who are married, working in private sector and having a masters degree
pysqldf("select count(*) as count_of_persons from adult where trim(marital_status) like 'Married%' and trim(workclass)='Private' and trim(education)='Masters'")
```

```python
# Obtaining sector level average,min and max ages using the respective methods
pysqldf("select workclass as sector,avg(age) as average_age,min(age) as minimum_age,max(age) as maximum_age from adult group by workclass order by 1 desc;")
```

```python
# Age distribution by country
pysqldf("select native_country,avg(age) as age_distribution from adult group by native_country order by 2 asc;")
```

```python
# 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'
adult['net_capital_gain'] = pysqldf("select capital_gain - capital_loss as net_capital_gain from adult;")
```
